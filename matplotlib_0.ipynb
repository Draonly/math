```python
from matplotlib import pyplot as plt

x = [1,2,3,1]
y = [1,3,0,1]

plt.plot(x,y)
plt.title('Triangle')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.ylim([-1,4])
plt.xlim([0,4])
plt.grid(True,color='r')
plt.show()
```

```python
# -*- coding: utf-8 -*-
from matplotlib import pyplot as plt
import numpy as np

x = [1,2,3,1]
y = [1,3,0,1]

x2 = np.array(x)+1
y2 = np.array(y)+1

plt.plot(x,y,'g',label=u'第一个', linewidth=5)
plt.plot(x2,y2,'c',label=u'第二个',linewidth=5)

plt.title(u'两个三角形')
plt.ylabel(u'Y轴')
plt.xlabel(u'X轴')

plt.legend()

plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.linspace(0,5,100)
y1 = np.sin(2*np.pi*x)
y2 = np.cos(2*np.pi*x)
y3 =y1+y2

fig = plt.figure()
ax1 = fig.add_subplot(221) 
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(212)

ax1.plot(x,y1,'g',label='one', linewidth=2)
ax1.set_title('one')
ax1.set_xlabel('x')
ax1.grid(True)

ax2.plot(x,y2,'c',label='two',linewidth=2)
ax2.legend()

ax3.plot(x,y3,'c',label='three',linewidth=2)

plt.show()
```
